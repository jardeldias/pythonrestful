```python
import time
import matplotlib.pyplot as plt
import pyotp
import qrcode

chave_mestre = 'abcdario'
codigo = pyotp.TOTP(chave_mestre)
print(codigo.now())

codigo_usuario = input('Código: ')
print(codigo.verify(codigo_usuario))

link = pyotp.TOTP(chave_mestre).provisioning_uri(name='lira', issuer_name='codigo_python')
meu_qrcode = qrcode.make(link)
meu_qrcode.save('qrcode.png')
```

```python
from http.server import BaseHTTPRequestHandler
import json

USERS = {
    'admin':{'role':'admin', 'password':'admin123'},
    'user':{'role':'user','password':'user123'}
}
def is_admin(username):
  user = USERS.get(username)
  return user and user['role'] == 'admin'

class MyHandler(BaseHTTPRequestHandler):
  def do_GET(self):
    if self.path == '/':
      self._send_response(200, {'message':'Bem-vindo à página inicial'})
    elif self.path == '/admin':
      username = self.headers.get('Username')
      if username and is_admin(username):
        self._send_response(200, {'message':'Bem-vindo ao painel de administração'})
      else:
        self._send_response(403, {'erro':'Acesso negado'})
    else:
      self._send_response(404, {'error':'Rota não encontrada'})
```
